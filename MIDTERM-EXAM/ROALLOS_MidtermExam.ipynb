```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heart_2020_cleaned.csv')
df
```

```python
plt.hist(df['BMI'])

plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Distribution of BMI')

plt.show()
```

**Question 1**: What is the average BMI within the dataset?

**ANSWER**: *The average BMI of the patients within the dataset is roughly in between 20 to 30, which means that most of the patients have Normal Weight (ranging from 18.5 to 24.9) to Overweight (ranging from 25 to 29.9).*  

```python
heartDisease = df['HeartDisease'].value_counts()
plt.pie(heartDisease.values, labels=heartDisease.index,autopct='%1.1f%%')

plt.title('"Do you suffer from a Heart Disease?"')

plt.show()
```

**Question 2**: How many of the recorded patients are suffering a heart disease?

**ANSWER**: *91.4% of the patients have responded 'No', while the remaining 8.6% have answered 'Yes' and therefore are experiencing a heart-related disease.*
