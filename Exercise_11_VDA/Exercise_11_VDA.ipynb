```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.50, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40])
y = np.array([1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')

plt.title('Correlation of Sample Data Sets')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')

plt.show()
```

# Interpretation

The graph illustrates a negative correlation between the two datasets. With the *r = -0.9475*, the datasets have a strong negative correlation.
