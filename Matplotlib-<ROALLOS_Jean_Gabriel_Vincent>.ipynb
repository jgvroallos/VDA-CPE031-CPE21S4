```python
#using pandas library to print out the dataframe

import pandas as pd

df = pd.read_csv('Tech Products Data.csv')
df
```

```python
# a. Device's date of release and sales

import matplotlib.pyplot as plt
import pandas as pd

x = df['Date of Release']
y = df['Price (USD)']

plt.figure(figsize=(12, 4)) # adjust figure size
plt.xticks(rotation=45, ha='right') # adjust label rotation

plt.title('Date of Release & Sales')
plt.xlabel('Product Release Date')
plt.ylabel('Price in USD')

plt.plot(x, y)

plt.show()
```

```python
# b. Get total sales

x = df['Date of Release']
y = df['Price (USD)']

plt.figure(figsize=(12, 4)) # adjust figure size
plt.xticks(rotation=45, ha='right') # adjust label rotation

plt.title('Date of Release & Sales')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.plot(x, y, color='r', marker='o', linestyle=':', linewidth=3, label='Sales')
plt.legend(loc='lower right')

plt.show()
```

```python
# c. read all product sales data and show it using a multi line plot

import matplotlib.pyplot as plt
import pandas as pd

x = df['Date of Release']
y = df['Price (USD)']

plt.figure(figsize=(12, 4)) # adjust figure size
plt.xticks(rotation=45, ha='right') # adjust label rotation

plt.title('Date of Release & Sales')
plt.xlabel('Product Release Date')
plt.ylabel('Price in USD')

plt.plot(x, y)

plt.show()
```

```python
# d. read each brand sales data and show it using a bar chart

x = df['Brand']
y = df['Price (USD)']

plt.figure(figsize=(12, 4))
plt.bar(x, y)

plt.title('Brand Sales')
plt.xlabel('Brand')
plt.ylabel('Sales in USD')

plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# e. read each device and show it using a histogram

histogramData = df['Country of Origin']

plt.xticks(rotation=45, ha='right')
plt.title('Most common Country of Origin')
plt.xlabel('Country')
plt.ylabel('Count')

plt.hist(histogramData)
plt.show()
```

```python
labels = df['Country of Origin']
labels.duplicated()
labels.drop_duplicates(inplace = True)
labels.reset_index(drop = True, inplace = True)
labels
```

```python
# f. Show each brand's Country of Origin

sizes =[14, 3, 2, 4, 6, 1]

plt.pie(sizes, labels=labels)

plt.legend(loc='upper left')
plt.title("Country of Origin")
plt.show()
```
